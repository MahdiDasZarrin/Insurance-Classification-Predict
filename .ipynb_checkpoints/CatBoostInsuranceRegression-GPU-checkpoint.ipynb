```python
data_train = pd.read_csv('train.csv')
data_test = pd.read_csv('test.csv')
```

```python
data_train['isTrain'] = True
data_test['isTrain'] = False
```

```python
traintest = pd.concat([data_train , data_test] , axis=0)
traintest.tail()
```

```python
import re
cat_pattern = re.compile("cat([1-9]|[1-9][0-9]|[1-9][0-9][0-9])$")
cont_pattern = re.compile("cont([1-9]|[1-9][0-9]|[1-9][0-9][0-9])$")
cat_col = sorted([cat for cat in traintest.columns if 'cat' in cat] , key=lambda s: int(s[3:]))
cat_col
```

```python
cat_index = [i for i in range(0, len(traintest.columns)) if cat_pattern.match(traintest.columns[i])]
cat_index
```

```python
cont_col = sorted([cont for cont in traintest.columns if 'cont' in cont] , key=lambda s: int(s[4:]))
cont_col
```

```python
cont_index = [i for i in range(0, len(traintest.columns)) if cont_pattern.match(traintest.columns[i])]
cont_index
```

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
```

```python
model = CatBoostRegressor(iterations=1000 , learning_rate=0.05 , depth=6 , eval_metric='MAE' , verbose=4)
```

```python
X = data_train.drop(['id','loss','isTrain'] , axis=1)
y = np.log(data_train['loss'])
X_train , X_test , y_train , y_test = train_test_split(X,y)
```

```python
model.fit(X_train , y_train , np.asarray(cat_index) - 1 , eval_set=(X_test , y_test))
```
